```python
# Import Dataset Sinubi
nama_file = "./Sinubi's Journey/Dataset SJ 01.xlsx"

data_df = pd.read_excel(nama_file, sheet_name='data')
```

```python
# Menampilkan Beberapa Data Awal pada Dataset Sinubi
data_df.head()
```

```python
# Membuat Subset Data untuk Mata Pelajaran "Pemrograman Python"
data_df_selected_1 = data_df[data_df['Mata Pelajaran'] == "Pemrograman Python"]
data_df_selected_1.head()
```

```python
# Menampilkan Statistik Deskriptif dari Variabel "Nilai"
# untuk "Mata Pelajaran" = "Pemrograman Python" pada Dataset Sinubi
data_df_selected_1['Nilai'].describe()
```

```python
# Membuat Subset Data untuk Mata Pelajaran "Machine Learning"
data_df_selected_2 = data_df[data_df['Mata Pelajaran'] == "Machine Learning"]
data_df_selected_2.head()
```

```python
# Menampilkan Statistik Deskriptif dari Variabel "Nilai"
# untuk "Mata Pelajaran" = "Machine Learning" pada Dataset Sinubi
data_df_selected_2['Nilai'].describe()
```

## *Multiple Swarm Plot* dari DataFrame

```python
# Membuat Multiple Swarm Plot
plt.figure(figsize=(5,3))                          # Ukuran Gambar (dalam inci)

sns.swarmplot(data = data_df,
              x = 'Mata Pelajaran',
              y = 'Nilai')

plt.title('Multiple Swarm Plot dari DataFrame')    # Judul Gambar
plt.xlabel('Nilai')                                # Keterangan untuk Sumbu X
plt.show()
```

```python
# Statistik Deskriptif untuk X1
X1 = data_df_selected_1['Nilai']

print('Min           : ', np.min(X1))
print('Max           : ', np.max(X1))
print('Q1            : ', np.quantile(X1,0.25))
print('Median (Q2)   : ', np.median(X1))
print('Q3            : ', np.quantile(X1,0.75))
print('IQR           : ', np.quantile(X1,0.75) - np.quantile(X1,0.25))

fence_atas = np.quantile(X1,0.75) + 1.5 * (np.quantile(X1,0.75) - np.quantile(X1,0.25))
if fence_atas > np.max(X1) : fence_atas = np.max(X1)

fence_bawah = np.quantile(X1,0.25) - 1.5 * (np.quantile(X1,0.75) - np.quantile(X1,0.25))
if fence_bawah < np.min(X1) : fence_bawah = np.min(X1)

print()
print('Fence/Extreme Atas  : ', fence_atas)
print('Fence/Extreme Bawah : ', fence_bawah)
```

```python
# Statistik Deskriptif untuk X2
X2 = data_df_selected_2['Nilai']

print('Min           : ', np.min(X2))
print('Max           : ', np.max(X2))
print('Q1            : ', np.quantile(X2,0.25))
print('Median (Q2)   : ', np.median(X2))
print('Q3            : ', np.quantile(X2,0.75))
print('IQR           : ', np.quantile(X2,0.75) - np.quantile(X2,0.25))

fence_atas = np.quantile(X2,0.75) + 1.5 * (np.quantile(X2,0.75) - np.quantile(X2,0.25))
if fence_atas > np.max(X2) : fence_atas = np.max(X2)

fence_bawah = np.quantile(X2,0.25) - 1.5 * (np.quantile(X2,0.75) - np.quantile(X2,0.25))
if fence_bawah < np.min(X2) : fence_bawah = np.min(X2)

print()
print('Fence/Extreme Atas  : ', fence_atas)
print('Fence/Extreme Bawah : ', fence_bawah)
```

### Membuat *Multiple Swarm Plot* menjadi Horizontal

```python
# Membuat Multiple Swarm Plot
plt.figure(figsize=(5,3))                          # Ukuran Gambar (dalam inci)

sns.swarmplot(data = data_df,
              y = 'Mata Pelajaran',
              x = 'Nilai')

plt.title('Multiple Swarm Plot dari DataFrame')    # Judul Gambar
plt.xlabel('Nilai')                                # Keterangan untuk Sumbu X
plt.show()
```

### Mengubah Warna pada *Multiple Swarm Plot*

```python
# Menyiapkan Warna Setiap Swarm
warna_swarm = ['green','orange']

# Membuat Multiple Swarm Plot
plt.figure(figsize=(5,3))                          # Ukuran Gambar (dalam inci)

sns.swarmplot(data = data_df,
              x = 'Mata Pelajaran',
              y = 'Nilai',
              palette = warna_swarm)

plt.title('Multiple Swarm Plot dari DataFrame')    # Judul Gambar
plt.xlabel('Nilai')                                # Keterangan untuk Sumbu X
plt.show()
```
