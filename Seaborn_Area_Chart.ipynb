```python
# Import Dataset Sinubi
nama_file = "/content/Dataset SJ 02.xlsx"

data_df = pd.read_excel(nama_file, sheet_name='data')
```

Dataset memiliki 4 Kolom dan 88 Baris

```python
# Membuat Subset Data untuk Mata Pelajaran "Pemrograman Python"
data_df_selected = data_df[data_df['Mata Pelajaran'] == "Pemrograman Python"]
data_df_selected.head()
```

## *Area Chart* dari DataFrame

```python
# Membuat Area Chart
plot = (
    so.Plot(data = data_df_selected,
            x = "Tanggal",
            y = "Jumlah Siswa Hadir")
    ).add(so.Area())

plot.label(title = "Area Chart dari DataFrame",     # Judul Gambar
           x = "Tanggal",                           # Keterangan untuk Sumbu X
           y = "Jumlah Siswa Hadir")                # Keterangan untuk Sumbu Y
```

### Mengubah Warna pada *Area Chart*

```python
# Membuat Area Chart
plot = (
    so.Plot(data = data_df_selected,
            x = "Tanggal",
            y = "Jumlah Siswa Hadir")
    ).add(so.Area(color = "red"))                   # Mengubah Warna Area Chart

plot.label(title = "Area Chart dari DataFrame",     # Judul Gambar
           x = "Tanggal",                           # Keterangan untuk Sumbu X
           y = "Jumlah Siswa Hadir")                # Keterangan untuk Sumbu Y
```
