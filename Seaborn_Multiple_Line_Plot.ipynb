## *Multiple Line Plot* dari DataFrame
sama dengan *Multiple Time-Series Plot*

```python
# Membuat Multiple Line Plot
plt.figure(figsize=(8,3))                        # Ukuran Gambar (dalam inci)

sns.lineplot(data = data_df,
             x = "Tanggal",
             y = "Jumlah Siswa Hadir",
             hue = "Mata Pelajaran")

plt.title('Multiple Line Plot dari DataFrame')   # Judul Gambar
plt.xlabel('Tanggal')                            # Keterangan untuk Sumbu X
plt.ylabel('Jumlah Kehadiran Siswa')             # Keterangan untuk Sumbu Y
plt.show()
```

### Mengubah Warna pada *Multiple Line Plot*

```python
# Membuat Multiple Line Plot
plt.figure(figsize=(8,3))                        # Ukuran Gambar (dalam inci)

sns.lineplot(data = data_df,
             x = "Tanggal",
             y = "Jumlah Siswa Hadir",
             hue = "Mata Pelajaran",
             palette = ["blue","red"])           # Mengubah Warna Line

plt.title('Multiple Line Plot dari DataFrame')   # Judul Gambar
plt.xlabel('Tanggal')                            # Keterangan untuk Sumbu X
plt.ylabel('Jumlah Kehadiran Siswa')             # Keterangan untuk Sumbu Y
plt.show()
```

### Mengubah Bentuk *Marker* pada *Multiple Line Plot*

```python
# Membuat Multiple Line Plot
plt.figure(figsize=(8,3))                        # Ukuran Gambar (dalam inci)

sns.lineplot(data = data_df,
             x = "Tanggal",
             y = "Jumlah Siswa Hadir",
             hue = "Mata Pelajaran",
             marker = "s")                       # Mengubah Bentuk Marker

plt.title('Multiple Line Plot dari DataFrame')   # Judul Gambar
plt.xlabel('Tanggal')                            # Keterangan untuk Sumbu X
plt.ylabel('Jumlah Kehadiran Siswa')             # Keterangan untuk Sumbu Y
plt.show()
```

### Mengubah Ukuran *Marker* pada *Multiple Line Plot*

```python
# Membuat Multiple Line Plot
plt.figure(figsize=(8,3))                        # Ukuran Gambar (dalam inci)

sns.lineplot(data = data_df,
             x = "Tanggal",
             y = "Jumlah Siswa Hadir",
             hue = "Mata Pelajaran",
             marker = ".",                       # Mengubah Bentuk Marker
             markersize = 8)                     # Mengubah Ukuran Marker

plt.title('Multiple Line Plot dari DataFrame')   # Judul Gambar
plt.xlabel('Tanggal')                            # Keterangan untuk Sumbu X
plt.ylabel('Jumlah Kehadiran Siswa')             # Keterangan untuk Sumbu Y
plt.show()
```
